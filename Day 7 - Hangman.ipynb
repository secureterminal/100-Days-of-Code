```python
# from nltk.corpus import words
# import random
# import nltk
# nltk.download('words')
import random
# all_words = words.words()
# # prints 236736
# print(len(word_list))

import all_words, stages

# load additional module
import pickle

with open('words.py', 'rb') as filehandle:
    # read the data as binary data stream
    all_word_list = pickle.load(filehandle)
# logo = stages.logo
stages = stages.stages


from all_words import logo
```

```python
# chosen_word = word_list[random.randint(0, len(word_list)-1)]
print(logo)
random.shuffle(all_word_list) # Not necessary
difficulty_level = input('Please select the lenght of words from 5 (easy mode) to 24 (legend mode)')
difficulty_check = difficulty_level.isdigit()

while not difficulty_check:
    print('Please input a valid number between 5 and 24')
    difficulty_level = input('Please select the lenght of words from 5 (easy mode) to 24 (legend mode)')
    difficulty_check = difficulty_level.isdigit()
    
    if difficulty_check:
        if int(difficulty_level) < 5 or int(difficulty_level) > 24:
            difficulty_check = False
    
word_list = []

for word in all_word_list:
    if len(word) == int(difficulty_level):
        word_list.append(word)
        
        
chosen_word = random.choice(word_list).lower()

print(chosen_word)


game_over = False
answer = []
guess_list = []

for a in range(len(chosen_word)):
    answer.append("_")

print(answer)

lives = 6
while not game_over and lives != 0:
    
    guess = input('\nPlease guess a letter\n').lower()
    print(guess)
    
    if guess not in guess_list:
        guess_list.append(guess)
        count = 0
        for a in chosen_word:
            if a == guess:
                answer[count] = a
            count+=1

        print(answer)
        if guess not in chosen_word:
            lives -= 1
            print(f'Wrong guess, sorry you lose a life and you have {lives} lives left')
            print(stages[lives])
            if lives == 0:
                print(f'Sorry you lost, the correct word is {chosen_word}')

        if '_' not in answer:
            print('User Wins')
            break
    else:
        print(f'You have already guessed "{guess}", please make another guess')
    
```
